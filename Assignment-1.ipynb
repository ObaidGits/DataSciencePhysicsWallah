##### 3. Compare and contrast mutable and immutable objects in Python with examples

In Python, objects are classified as mutable or immutable based on whether their values can change after creation.

Mutable Objects:
Can be modified after creation.
Examples: Lists, dictionaries, sets.

Immutable Objects:
Cannot be changed after creation.
Examples: Strings, tuples, integers, floats.

Comparison:
Mutability: Mutable objects can be altered in place, while immutable objects require creating a new object for changes.
Performance: Immutable objects can be more memory efficient and safer for concurrency, as their state cannot change unexpectedly.

##### 5) Explain the concept of type casting in python with examples.

Type casting in Python refers to converting one data type into another. It allows developers to change the data type of a variable or value to match the desired format in a particular context.

Types of Type Casting:

Implicit Type Casting (Automatic):
Python automatically converts a smaller data type to a larger data type during operations to avoid data loss.
###### Example Code Below:
> num_int = 5       # Integer

> num_float = 3.5   # Float

> result = num_int + num_float  # int is implicitly converted to float

> print(result)  # Output: 8.5 (float)

Explicit Type Casting (Manual):
Developers manually convert one data type into another using built-in functions like int(), float(), str(), etc.
###### Example Code Below:
###### Integer to String
> num = 10

> num_str = str(num)

> print(num_str)  # Output: '10' (string)

###### Example Code Below:
###### Float to Integer
> num_float = 5.67

> num_int = int(num_float)  # Truncates the decimal part

> print(num_int)  # Output: 5

###### Common Type Casting Functions:
> int(): Converts to integer.

> float(): Converts to float.

> str(): Converts to string.

> st(), tuple(), set(): Convert to list, tuple, or set.

> Type casting ensures compatibility and flexibility in operations involving different data types.

##### 7. Describe different types of loops in Python with their use cases with examples.

```
Python has two main types of loops:

1. for loop:
Iterates over a sequence (like a list, tuple, or string).

Use case: When you know the number of iterations beforehand.

python
Copy code
for i in range(3):
    print(i)  # Output: 0, 1, 2
2. while loop:
Repeats as long as a condition is True.

Use case: When the number of iterations is unknown and based on a condition.

python
Copy code
i = 0
while i < 3:
    print(i)
    i += 1  # Output: 0, 1, 2
Loop Control:
break: Exit loop early.
continue: Skip current iteration.
Each loop is used based on iteration needs and conditions.
```
